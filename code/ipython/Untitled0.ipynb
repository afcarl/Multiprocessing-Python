```python
# build decks
import numpy
def build_decks():
    decks = []
    for i in range(5):
        deck = numpy.arange(1,53,1)
        numpy.random.shuffle(deck)
        decks.append(deck)
    return decks

decks = build_decks()
print decks
```

```python
def single_sort_1(decks):
    sorted_decks = []
    for deck in decks:
        sorted_deck = sorted(deck)
        sorted_decks.append(sorted_deck)
    return sorted_decks

print "\nsingle_sort_1"
print single_sort_1(decks)
```
